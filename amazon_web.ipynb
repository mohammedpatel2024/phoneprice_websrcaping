```python
url="https://www.amazon.in/s?k=phones"
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
Header=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/137.0.0.0 Safari/537.36','Accepted-Language':'en-US,en;q=0.5'})
```

```python
webpage=requests.get(url,headers=Header)
```

```python
webpage
```

```python
webpage.content
```

```python
type(webpage.content)
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
links=soup.find_all("a",attrs={"class":"a-link-normal s-line-clamp-2 s-line-clamp-3-for-col-12 s-link-style a-text-normal"})
```

```python
links
```

```python
link=links[0].get('href')
```

```python
product_link="https://amazon.com"+link
```

```python
product_link
```

```python
import urllib.parse

track_link = links[0].get("href")

encoded = track_link.split("url=")[1].split("&")[0]
real_path = urllib.parse.unquote(encoded)
#percectperfect 👌 
product_link = "https://www.amazon.in" + real_path
print(product_link)
```

```python
new_webpage=requests.get(product_link,headers=Header)
```

```python
new_webpage
```

```python
new_soup=BeautifulSoup(new_webpage.content,"html.parser")
```

```python
new_soup
```
